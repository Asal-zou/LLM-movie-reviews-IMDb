```python
from transformers import pipeline

# Initialize the pipeline for sentiment analysis with DistilBERT
pipe = pipeline("text-classification", model="distilbert-base-uncased")
```

```python
from datasets import load_dataset
import numpy as np
import pandas as pd

ds = load_dataset('imdb')
ds_train = pd.DataFrame(ds['train'])
ds_test = pd.DataFrame(ds['test'])
```

```python
from datasets import Dataset, DatasetDict
# assign the splits
train = Dataset.from_pandas(ds_train)
test = Dataset.from_pandas(ds_test)
# reconstruct both datasets into a Dataset Dict object
new_ds = DatasetDict(
    {
        'train': train,
        'test': test
    }
)
# view the resulting dataset dict object
new_ds
```

Preparing Data for Inference:
You want to prepare a list of text data from your dataset ds for inference

```python
# Preparing the data
data = ds['test']['text'][:5]  # Extracting the first 5 text entries from the test dataset
```

Making Predictions with the Pipeline:
Next, you use the pipeline pipe to process the data and generate predictions:

```python
# Making predictions
preds = pipe(data)
```

```python
# Inspecting predictions
for i, pred in enumerate(preds):
    print(f"Text {i+1}: {data[i]}")
    print(f"Predicted Sentiment: {pred['label']} with confidence {pred['score']:.4f}\n")
```
